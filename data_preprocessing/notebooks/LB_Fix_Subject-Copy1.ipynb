```python
# Loop over the trials
df_lb_trial['offsetFixation_idx'] = -9999
for index, row in df_lb_trial.iterrows():
    target_offset = row['target_offset']
    
    beforePreviousFix = df_lb.loc[df_lb.ts < target_offset]
    lastfixation = beforePreviousFix.iloc[-1]
    nextfixation = df_lb.loc[beforePreviousFix.index[-1] + 1]
    df_lb_trial.loc[index, "offsetFixation_idx"] = beforePreviousFix.index[-1]
```
