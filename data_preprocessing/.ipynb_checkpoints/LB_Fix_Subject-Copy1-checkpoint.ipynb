```python
df_lastFix = pd.DataFrame([df_lb.iloc[df_lb.index.get_loc(t, method='nearest')] for t in df_lb_trial.index])
```

```python
# df_lastFix["x"] = pd.to_numeric(df_lastFix["x"],downcast='integer')
df_lastFix["x"] = df_lastFix['x'].apply(lambda x: int('%.0f' % x))
df_lastFix["y"] = df_lastFix['y'].apply(lambda x: int('%.0f' % x))
df_lastFix = df_lastFix.reset_index()
df_lb_trial = df_lb_trial.reset_index()
df_lastFix["disp"] = pd.to_numeric(df_lastFix["disp"])
df_lastFix.head()
```

```python
#SETUP COLORS FOR THE GRAPHS
df_lastFix['color'] = pd.Series(dtype='int')
df_lastFix.loc[(df_lastFix.disp < 50), 'color'] = "Green"
df_lastFix.loc[(df_lastFix.disp > 50) & (df_lastFix.disp < 100), 'color'] = "Yellow"
df_lastFix.loc[(df_lastFix.disp >= 100) & (df_lastFix.disp < 500), 'color'] = "Red"
```

```python
# Caluclate The Euclidan Distance

x1 = df_lastFix.x
# x1
x2 = df_lb_trial.targetX
# x2
y1 = df_lastFix.y
# y1
y2 = df_lb_trial.targetY
# y2
d2 = np.square( x2 - x1 )  + np.square( y2 - y1 ) 


distances = np.sqrt( d2 )
df_lastFix['distance'] = distances
```

```python
fig, ax = plt.subplots(1, figsize=(9,7))
plt.scatter(df_lb_trial.targetX, df_lb_trial.targetY, marker='o', c=df_lastFix.color, s=df_lastFix.disp)
plt.gca().invert_yaxis()
plt.title('Participant number - 1 last fixations centroid grupped around the 56 targets in the Large Gird Paradigm, showin in the different colours the Dispersion in pixels \n Dispersion less then 50 - GREEN \n Dispersion between 50 and 100 PIXELS - YELLOW \n Dispersion more then 100 PIXELS - RED')
plt.xlabel('x coordinates in pixel units')
plt.ylabel('y coordinates in pixel units')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def distanceToTargetPlot(frows,lrows):
    fig, ax = plt.subplots(1, figsize=(16,10))
    plt.gca().invert_yaxis()
    t = df_lastFix.loc[frows:lrows]
    t1 = df_lb_trial.loc[frows:lrows]

    point1 = [t1.targetX, t1.targetY]
    point2 = [t.x, t.y]

    x_values = [point1[0], point2[0]]

    y_values = [point1[1], point2[1]]



#     plt.plot(t.x, t.y, color='pink', linestyle = 'dashed')
    plt.plot(x_values, y_values)
    plt.scatter(t1.targetX, t1.targetY, marker='+', c='Black', s=100)
    plt.scatter(t.x, t.y, marker='^', s=100)
    plt.title('LABVANCED DATA: Patricipant 1 with the fixation centroid grupped around the 56 target in the Large Gird Paradigm',fontsize=20)
    plt.xlabel('x coordinates in pixel units',fontsize=12)
    plt.ylabel('y coordinates in pixel units',fontsize=12)
    plt.grid(True)
    plt.show()
```

```python
distanceToTargetPlot(0,56)
```
